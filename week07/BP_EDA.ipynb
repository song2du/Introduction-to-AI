```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/송이두/Desktop/가천대/2-1/인공지능개론/과제/data/BP_data.csv')
```

# Count Plot

```python
sns.countplot(x='Sex', data=data)
plt.title('Freqency of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

# BoxPlot

```python
sns.boxplot(y=data['Physical_activity'])  # y축에 컬럼 데이터 지정
plt.title('Box Plot of Physical activity')
plt.ylabel('Value')
plt.show()
```

```python
sns.boxplot(x='Sex', y='Level_of_Hemoglobin', data=data)
```

# ViolinPlot

```python
sns.violinplot(x='Sex', y='Level_of_Hemoglobin', data=data)
plt.show()
```

# Hist

```python
sns.histplot(x='Age', data=data, kde=True)
```

```python
plt.hist(data['BMI'], bins=5, edgecolor='black')  # bins는 막대의 개수, edgecolor는 막대 테두리 색상
plt.title('Histogram of BMI')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Heatmap

```python
df_corr=data.corr()

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

# 남여에 따른 스트레스 차이

```python
sns.violinplot(x='Sex', y='Level_of_Stress', data=data)
plt.show()
```

# BMI와 흡연 여부에 따른 고혈압 여부

```python
sns.boxplot(x='Blood_Pressure_Abnormality', y='BMI', hue='Smoking', data=data)
plt.title('BMI Distribution by Blood Pressure Abnormality and Smoking Status')
plt.show()
```
